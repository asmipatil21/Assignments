
In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
import seaborn as sns
%matplotlib inline
```

```python
iris = datasets.load_iris()
```

```python
fig = plt.figure(figsize = (8,8))
ax = Axes3D(fig, elev = -150, azim = 110)
X = PCA(n_components = 3).fit_transform(iris.data)
Y = iris.target
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c = Y, cmap = plt.cm.rainbow_r)
ax.set_title("3D plot")
ax.set_xlabel("1st direction")
ax.set_ylabel("2nd direction")
ax.set_zlabel("3rd direction")
plt.show()
```
